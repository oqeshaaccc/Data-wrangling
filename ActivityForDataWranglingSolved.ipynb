```python
import pandas as pd
df = pd.read_csv('auto-mpg.csv')
df.head()
```

```python
#Drop column Origin from the dataset.
df.drop(columns =['origin'], axis=1, inplace=True)
df.head()
```

```python
'''
Find out all the missing values in the dataset and replace it with its most appropriate replacement. 
Finding out missing values:  We can use .sum() method after applying .isnull() this will return sum of missing values within each columns in the dataframe.
import pandas as pd
 df = pd.read_csv('IMDB-Movie-Data.csv')
df.isnull().sum()
'''
#using mean to replace
import numpy as np
df['mpg'].replace(np.nan, df['mpg'].mean(), inplace= False)
df['cylinders'].replace(np.nan, df['cylinders'].mean(), inplace= False)
df['displacement'].replace(np.nan, df['displacement'].mean(), inplace= False)
df['horsepower'].replace(np.nan, df['horsepower'].mean(), inplace= False)
df['weight'].replace(np.nan, df['weight'].mean(), inplace= False)
df['acceleration'].replace(np.nan, df['acceleration'].mean(), inplace= False)
df['model year'].replace(np.nan, df['model year'].mode()[0], inplace= False)
df['car name'].replace(np.nan, df['car name'].mode()[0], inplace= False)
df['CAR Number'].replace(np.nan, df['CAR Number'].mode()[0], inplace= False)

df
```

```python
#Find and remove duplicate entries for the column ‘ CAR Number’
df['CAR Number'] = df['CAR Number'].drop_duplicates()
df['CAR Number']
```

```python
'''
In our dataset, the fuel consumption column is "mpg" and is represented by mpg (miles per gallon) unit. 
Assume we are developing an application in a country that accepts fuel consumption with the L/100km standard.
change the name of the column to "L/100km". 
'''
df['L/100km'] = df['mpg'] * 0.425144
df.drop(columns=['mpg'],axis=1, inplace= True)
df
```

```python
'''
Normalize the column “Weight” so that the values range from 0 to 1 
Normalization is the process of transforming values of several variables into a similar range. Typical 
normalizations include scaling the variable so the variable average is 0, scaling the variable so the variance is 1,
or scaling variable so the variable values range from 0 to 1
'''
df['weight'] = (df['weight']-df['weight'].mean())/df['weight'].std()#used zscore
df
```

```python
'''
In our dataset, "horsepower" is a real valued variable ranging from 48 to 288, 
it has 57 unique values. What if we only care about the price difference between cars with 
high horsepower, medium horsepower, and little horsepower (3 types)? 
Can we rearrange them into three ‘bins' to simplify analysis?
'''
#We will use the Pandas method 'cut' to segment the 'horsepower' column into 3 bins

import numpy as np 
groups=['high', 'medium','little']
horsepower_category_bins= np.linspace(df['horsepower'].min(), df['horsepower'].max(),4)
df['horsepower_category_bins'] = pd.cut(df['horsepower'], horsepower_category_bins, labels = groups, include_lowest = True)
df
```

```python
#Use a histogram to visualize the distribution of bins we created above.
df['horsepower_category_bins'].hist()
```
